```python
# Useful imports.
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats as stats
from scipy.stats import binom

# Set the random seed.
np.random.seed(8302)
```

<h4> Binomial distribution </h4>

The binomial distribution represents the number of successes in $n$ independant trials with a probability of success $p$ (and a probability of failure $(1-p)$).

$$P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}$$

$$P(X \leq k) = \sum_{y \leq k} \binom{n}{y} p^y (1-p)^{n-y}$$

NOTE: a binomial distribution of $n=1$ trials is a Bernoulli distribution.

```python
# Visualize the binomial distribution.
def visualize_distribution(x, y, xlabel, title):
    plt.bar(x, y)
    plt.xlabel(xlabel)
    plt.ylabel('Probability')
    plt.title(title)
    plt.show()
```

```python
# Create a binomial distribution.
n = 10 # number of trials
p = 0.5 # probability of success
x = np.arange(0, n+1) # possible outcomes
y = binom.pmf(x, n, p) # binomial probability mass function
visualize_distribution(x, y, 'Number of successes', 'Binomial Distribution')
```

```python
# Simulate a binomial distribution.
n = 5
p =  0.4
simulations = 10
y = np.random.binomial(n, p, simulations)
print("Results of simulations: ", y)
```

<h4> Categorical distribution </h4>

One trial with probability $p$.

$P(X = i) = p_i$ for $i \in {1,2,...,k}$ where $\sum_{i=1}^k p_i = 1$.

```python
# Categorial Distribution.
categories = ['A', 'B', 'C', 'D']
probabilities = [0.4, 0.2, 0.3, 0.1]
probabilities_sum = np.sum(probabilities) # should be 1
print("Probabilities should sum at 1: ", round(probabilities_sum, 5))
y = probabilities
visualize_distribution(categories, y, 'Categories', 'Categorical Distribution')
```

```python
# Simulate a categorical distribution.
simulations = 10
y = np.random.choice(categories, simulations, p=probabilities)
print("Results of simulations: ", y)
```
